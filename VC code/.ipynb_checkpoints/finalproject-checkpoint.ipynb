````python
def Sex_recode (series):
    if series == " Female":
        return 0
    if series == " Male":
        return 1
    if series == " Both sexes":
        return 2
    
crude["sexR"] = crude['Sex'].apply(Sex_recode)
crude.drop('Sex', axis = 1, inplace = True)
````
